```python
import os

files = [f for f in os.listdir('.') if os.path.isfile(f)]

print("Files in the current directory:")
for file in files:
    print(file)

```

```python
from pathlib import Path

files = [f.name for f in Path('.').iterdir() if f.is_file()]

print("Files in the current directory:")
for file in files:
    print(file)
```
